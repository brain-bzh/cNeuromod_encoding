```python
import os
import pandas as pd
import numpy as np
from statsmodels.multivariate.manova import MANOVA
```

```python
models = ['STG', 'wholebrain']
subjects = ['sub-01', 'sub-02', 'sub-03', 'sub-04', 'sub-05', 'sub-06']
convs = ['conv4', 'baseline']

annotation_path = './indices_friends_av - annotations.tsv'
df = pd.read_csv(annotation_path, sep='\t')

tags = ['tag_Talking', 'tag_Music', 'tag_laugh', 'tag_woman_speak', 'tag_man_speak', 'tag_Applause', 'tag_Car']
```

```python
df_tags = df.loc[(df['conv'].isna()) & (df['name'].isin(tags))]
df_r2 = df.loc[df['name'].isna()]
df_seasons = df.loc[df['name'].isin(['s01', 's02', 's03', 's04'])]
```

```python
for i, row in df_r2.iterrows():
    model = row['model']
    subject = row['subject']
    conv = row['conv']
    print(subject, model, conv)
    model_df = row.to_frame().T.drop(['model', 'subject', 'conv'], axis=1)
    model_df['name'] = 'model_r2'
    df_model_tags = pd.concat([model_df, df_tags, df_seasons]).set_index('name').drop(['model', 'subject', 'conv'], axis=1).T
    #print(df_model_tags.corr())
    print(df_model_tags)
    
    
    formula = ''
    for tag in tags:
        #formula += f'{tag} + '
        formula = tag + 's01 + s02 + s03 ' + '~ model_r2 + s04'
        print(formula)
        fit = MANOVA.from_formula(formula, data=df_model_tags)
        print(fit.mv_test())
    

    
```
